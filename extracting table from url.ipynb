```python
import pandas as pd

# FluWatch report page URL
url = "https://www.canada.ca/en/public-health/services/surveillance/respiratory-virus-detections-canada/2018-2019/respiratory-virus-detections-isolations-week-01-ending-january-5-2019.html"

# Read all HTML tables from the webpage
tables = pd.read_html(url)

# Table 2 is usually the second table, so index = 1
df = tables[1]

# Display the DataFrame
df.head()

```

```python
df.columns[0]
```

```python
df = df.rename(columns={
    df.columns[0]: "Province/Territory"
})

df
```

```python
df.to_csv("fluwatch_table2.csv", index=False)
```
